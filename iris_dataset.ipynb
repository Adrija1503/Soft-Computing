```
# Normalising Data
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target


# Convert to DataFrame for easier manipulation
df = pd.DataFrame(X, columns=iris.feature_names)
df['target'] = y

# Split the dataset into Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)


# Initialize the MinMaxScaler
scaler_minmax = MinMaxScaler()

# Fit the scaler on the training data and transform the training data
X_train_scaled_minmax = scaler_minmax.fit_transform(X_train)

# Transform the test data using the same scaler (without fitting again)
X_test_scaled_minmax = scaler_minmax.transform(X_test)

# Convert to DataFrame for easier inspection (optional)
df_train_scaled_minmax = pd.DataFrame(X_train_scaled_minmax, columns=iris.feature_names)
df_train_scaled_minmax['target'] = y_train

df_test_scaled_minmax = pd.DataFrame(X_test_scaled_minmax, columns=iris.feature_names)
df_test_scaled_minmax['target'] = y_test

# Display the first few rows of the scaled training and test data
print("Min-Max Scaled Training Data:\n", df_train_scaled_minmax.head())
print("Min-Max Scaled Test Data:\n", df_test_scaled_minmax.head())

# Verify the normalization
print("\nMinMaxScaler - Training Data Statistics:\n", df_train_scaled_minmax.describe())
print("\nMinMaxScaler - Test Data Statistics:\n", df_test_scaled_minmax.describe())
```

```
# LogisticRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
classifier = LogisticRegression()
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

# Summary of the predictions made by the classifier
print(classification_report(y_test, y_pred))
cm=confusion_matrix(y_test, y_pred)
print(confusion_matrix(y_test, y_pred))
# Accuracy score
from sklearn.metrics import accuracy_score
print('accuracy is',accuracy_score(y_pred,y_test))
```

```
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True)
plt.xlabel('predicted')
plt.ylabel('truth')
```

```
# K-Nearest Neighbours
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors=8)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

# Summary of the predictions made by the classifier
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
# Accuracy score
from sklearn.metrics import accuracy_score
print('accuracy is',accuracy_score(y_pred,y_test))
```
